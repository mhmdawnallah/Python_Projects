```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
movies.head()
```

```python
sns.set_style('darkgrid')
#g = sns.FacetGrid(movies,row='Genre',hue='Genre')
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
```

```python
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
g = g.map(plt.scatter,'CriticalRating','AudienceRating')
```

```python
#can be populated with any type of chart. Example:Histograms
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
g = g.map(plt.hist,'BudgetMillions')
```

```python
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5,edgecolor='black')
g = g.map(plt.scatter,'CriticalRating','AudienceRating',**kws)
```

---

```python
#Controlling Axes and Adding Diagonals
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5,edgecolor='black')
g = g.map(plt.scatter,'CriticalRating','AudienceRating',**kws)
g.set(xlim=(0,100),ylim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='--')
g.add_legend()
```
